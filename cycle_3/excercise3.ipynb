```python
x1 = [2,4,6,8]
y1 = [7, 8, 4,12]

x2 = [3,5,7,9]
y2 = [6,5,8,14]

fig,sp = plt.subplots(2)
fig.suptitle("Mutliple plots")
sp[0].plot(x1,y1, color='r')
sp[1].plot(x2,y2, color='g')
plt.show()
```

```python
courses=['Java','Python', 'PHP','JavaScript','C#', 'C++']
popularity=[22.2,17.6,8.8,8,77,6.7]

plt.bar(courses,popularity,width=0.4)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
plt.barh(courses, popularity, color='red')
plt.ylabel('Programming Languages')
plt.xlabel('Popularity')
plt.show()
```

```python
c = ['red','green','blue','maroon','blue','yellow']

plt.bar(courses, popularity,color=c)
plt.xlabel('Programming Languages')
plt.ylabel('Popularity')
plt.show()
```

```python
Women = [22, 30, 35, 35, 26]
Men = [25, 32, 30, 35, 29]
  
n=len(Men)
r = np.arange(n) 
width = 0.25
  
plt.bar(r , Men, color = '#00ad48', width = width, edgecolor = 'black',label='Men') 
plt.bar(r + width, Women, color = '#bf0f3b',width = width, edgecolor = 'black', label='Women') 

plt.xlabel("Person") 
plt.ylabel("Scores") 
plt.title("Scores by person") 
  

plt.xticks(r + width/2,['G1','G2','G3','G4','G5']) 
plt.legend() 
  
plt.show() 
```

```python
y = np.array([22.2 ,  17.6 ,  8.8,  8 ,  77  ,6.7])
languages = ["Java" ,  "Python" , "PHP" , "JavaScript" ,"C#" ,"C++"]

plt.pie(y, labels = languages) 
#Pie chart is drawn from y and the corresponding labels are selected from the list:- languages

plt.show()
```
